*How do the plots differ:*
- For the actual values plot
- For their derivatives plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_derivative(func, value):
    smaaaaaaaaal_value = 1e-5
    old = func(value)
    new = func(value + smaaaaaaaaal_value)
    height = new - old
    slope = (height / smaaaaaaaaal_value).round(3)
    return slope
```

```python
def plot(func, data):
    fig, axes = plt.subplots(1, 2, figsize=(20, 5))

    axes[0].plot(data, func(data))
    axes[0].set_title("Value plot")

    axes[1].plot(data, get_derivative(func, data))
    axes[1].set_title("Derivative plot");
```

## Linear

```python
def linear(x):
    return 2 * x
```

```python
data = np.arange(-10, 10)
plot(linear, data)
```

## Square

```python
def square(x):
    return x**2
```

```python
data = np.arange(-10, 10)
plot(square, data)
```

## Log

```python
def log(x):
    return np.log(x)
```

```python
data = np.arange(-10, 10)
plot(log, data)
```

## Some Quadratic!

```python
def quad(x):
    return x**4
```

```python
data = np.arange(-100, 100)
plot(quad, data)
```

## Sigmoid

```python
def sig(x):
    return 1 / (1 + np.exp(-x))
```

```python
data = np.linspace(-10, 10, 100)
plot(sig, data)
```

## TanH

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
data = np.linspace(-10, 10, 100)
plot(tanh, data)
```

## ReLU

```python
def relu(x):
    return np.maximum(0, x)
```

```python
data = np.linspace(-10, 10, 100)
plot(relu, data)
```

## Leaky ReLU

```python
def leaky_relu(x, alpha=0.2):
    return np.maximum(alpha * x, x)  # Or np.where(x > 0, x, alpha * x) for same vibe
```

```python
data = np.linspace(-10, 10, 100)
plot(leaky_relu, data)
```
